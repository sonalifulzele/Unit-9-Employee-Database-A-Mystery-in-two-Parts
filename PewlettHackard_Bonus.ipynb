```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Connect the Pewlett Hackard database on Postgres

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://User:Password@localhost/PewlettHackard')
```

### Query the database to get Title and Salary data

```python
data_salary=[]
conn = engine.connect()
result = conn.execute("select t.title as title, AVG(s.salary) as avgsal from salaries s join titles t on s.emp_no = t.emp_no group by t.title")
for row in result:
    data_salary.append(row)
conn.close()
data_salary
```

### Plot the bargraph

```python
x_axis, y_axis = zip(*data_salary)
plt.xlabel("Average Salary")
plt.ylabel("Titles")
plt.title("Average salaries in Pewlett Hackard by Title")

plt.barh(x_axis, y_axis, height=0.5)

plt.tight_layout()
plt.show()
```
